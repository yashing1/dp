```python
import numpy as np
from skimage.io import imread
from skimage.color import rgb2gray
from skimage import data, img_as_float, img_as_ubyte, exposure, io, color
from PIL import Image, ImageEnhance, ImageFilter
from scipy import ndimage, misc
import matplotlib.pyplot as pylab
import cv2
```

```python
def plot_image(image, title=""):
    pylab.title(title, size=10), pylab.imshow(image)
    pylab.axis('off')
```

```python
def plot_hist(r,g,b,title=""):
    r,g,b=img_as_ubyte(r),img_as_ubyte(g),img_as_ubyte(b)
    pylab.hist(np.array(r).ravel(),bins=256, range=(0,256),color='r',alpha=0.3)
    pylab.hist(np.array(g).ravel(),bins=256, range=(0,256),color='g',alpha=0.3)
    pylab.hist(np.array(b).ravel(),bins=256, range=(0,256),color='b',alpha=0.3)
    pylab.xlabel('Pixel Values', size=20), pylab.ylabel('Frequency',size=20)
    pylab.title(title,size=10)

im=Image.open('images/bird.jpg')
im_r,im_g,im_b=im.split()
pylab.style.use('ggplot')
pylab.figure(figsize=(15,5))
pylab.subplot(121), plot_image(im, 'Orginal Image')
pylab.subplot(122),plot_hist(im_r,im_g,im_b,'Histogram of Image')
pylab.show()
```

```python
im=Image.open('images/elephant2.jpeg')
im_r,im_g,im_b=im.split()

```

```python
imlog=im.point(lambda i: int(255*np.log(1+i/255)))
im_rlog,im_glog,im_blog=imlog.split()
pylab.style.use('ggplot')
pylab.figure(figsize=(15,5))
pylab.subplot(121), plot_image(imlog, 'Log Transofrmation of Original Image')
pylab.subplot(122),plot_hist(im_rlog,im_glog,im_blog,'Histogram of Log Image')
pylab.subplot(221), plot_image(im, 'Orginal Image')
pylab.subplot(222),plot_hist(im_r,im_g,im_b,'Histogram of Image')
pylab.show()



```

```python
import numpy as np
from skimage.io import imread
from skimage.color import rgb2gray
from skimage import data, img_as_float, img_as_ubyte, exposure, io, color
from PIL import Image, ImageEnhance, ImageFilter
from scipy import ndimage, misc
import matplotlib.pyplot as pylab
import cv2
```

```python
def plot_image(image, title=""):
    pylab.title(title, size=10), pylab.imshow(image)
    pylab.axis('off')
```

```python
im=img_as_float(imread('images/bird.jpg'))
gamma=2.5
imgamma=im**gamma
pylab.style.use('ggplot')
pylab.figure(figsize=(15,5))
pylab.subplot(121)
plot_hist(im[...,0],im[...,1],im[...,2]) # histogram for RGB(Input)
pylab.subplot(122)
plot_hist(imgamma[...,0],imgamma[...,1],imgamma[...,2]) # histogram for RGB(Input)
pylab.show()

pylab.figure(figsize=(15,5))
pylab.subplot(121)
plot_image(im,'Original Image')
pylab.subplot(122)
plot_image(imgamma,'output')
pylab.show()
```
